```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
%matplotlib inline
```

```python
df = pd.read_csv("heart.csv")
df.head(20)
```

```python
print("The shape of the dataset is : ", df.shape)
```

```python
df.describe()
```

```python
df.isnull().sum().sort_values()
```

```python
s_imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
s_imputer = s_imputer.fit(df)
df = s_imputer.transform(df)
```

```python
df.shape
```

```python
data = pd.DataFrame(df)
data
```

```python
data.isnull().sum().sort_values()
```

```python
from sklearn.model_selection import train_test_split

x = data.iloc[: , :13]
y = data.iloc[: , -1]
x_train, x_test , y_train , y_test = train_test_split(x, y , test_size= .2 , random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logisticRegModel = LogisticRegression(max_iter=2000)
logisticRegModel.fit(x_train , y_train)
y_predict = logisticRegModel.predict(x_test)
accuracy1 = accuracy_score(y_test , y_predict)

print('The accuracy score of the Logistic Model is ',accuracy1)
```

```python
from sklearn.svm import SVC

SVCModel = SVC(kernel='linear')
SVCModel.fit(x_train, y_train)
y_predict = SVCModel.predict(x_test)
accuracy2  = accuracy_score(y_test, y_predict)
print('The accuracy score of the SVC Model is ', accuracy2)
```

```python
from sklearn.neural_network import MLPClassifier

NNModel = MLPClassifier(hidden_layer_sizes=(100,), activation='relu', solver='adam', max_iter=1000)
NNModel.fit(x_train , y_train)
y_predict = NNModel.predict(x_test)
accuracy3 = accuracy_score(y_test, y_predict)
print('The accuracy score of the Logistic Model is ',accuracy3)
```
